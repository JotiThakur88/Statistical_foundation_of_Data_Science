```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("student_rating_dataset_100.csv")  # replace with actual file path
df.head()
```

```python
# Filter visible minorities
minorities = df[df['visible_minority'] == 1]

# Percentage of minorities who are tenured
percent_tenured_minority = minorities['tenure'].mean() * 100
percent_tenured_minority
```

```python
# Group by tenure and calculate mean & std
age_stats = df.groupby('tenure')['age'].agg(['mean', 'std'])
age_stats
```

```python
# Histogram
plt.hist(df['age'], bins=10, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Professors')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Count of genders
gender_counts = df['gender'].value_counts()

# Vertical bar
plt.bar(gender_counts.index, gender_counts.values, color=['blue','pink'])
plt.title('Number of Professors by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Horizontal bar
plt.barh(gender_counts.index, gender_counts.values, color=['blue','pink'])
plt.title('Number of Professors by Gender (Horizontal)')
plt.xlabel('Count')
plt.ylabel('Gender')
plt.show()
```
