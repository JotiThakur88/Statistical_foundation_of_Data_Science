```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ---------------------------------------------------------------
# Q2. Mean, Median, Minimum and Maximum for 'student_rating'
# ---------------------------------------------------------------

# Step 1: Import dataset
import pandas as pd

df = pd.read_csv("teacher_ratings.csv")
# Step 2: Display first 5 rows to confirm import
print("First 5 rows of the dataset:")
print(df.head())

```

```python
# Step 3: Calculate summary statistics for student_rating
mean_rating = df["student_rating"].mean()
median_rating = df["student_rating"].median()
min_rating = df["student_rating"].min()
max_rating = df["student_rating"].max()
# Step 4: Print results
print("\nQ2. Student Rating Summary:")
print(f"Mean = {mean_rating:.2f}")
print(f"Median = {median_rating:.2f}")
print(f"Minimum = {min_rating:.2f}")
print(f"Maximum = {max_rating:.2f}")
```

```python
print("\nQ3. Descriptive Statistics Table:")
print(df.describe())
```

```python
# Q4. Histogram of Beauty Variable

plt.figure(figsize=(8,5))
plt.hist(df["beauty"], bins=10, edgecolor='black', color='skyblue')
plt.title("Distribution of Beauty Ratings")
plt.xlabel("Beauty Score")
plt.ylabel("Frequency")
plt.show()

print("\nQ4. Comment:")
print("ðŸ‘‰ The beauty ratings appear roughly normally distributed, with most values between 5 and 8.")
```

```python
# Q5. Does average beauty score differ by gender?
# ---------------------------------------------------------------
print("\nQ5. Average Beauty Score by Gender:")
beauty_stats = df.groupby("gender")["beauty"].agg(["mean", "std"])
print(beauty_stats)

print("\nComment:")
print("The average beauty score is compared between males and females using mean and standard deviation.")
```

```python
# ---------------------------------------------------------------
# Q6. Percentage of Tenured Professors by Gender (Warning-Free)
# ---------------------------------------------------------------
tenure_percent = (
    df.groupby(["gender", "tenured"], group_keys=False)  # added group_keys=False
    .size()
    .groupby(level=0)
    .apply(lambda x: 100 * x / x.sum())
    .unstack()
    .fillna(0)
)

print("\nQ6. Percentage of Tenured Professors by Gender (%):")
print(tenure_percent)

print("\nComment:")
print("Tenure status appears fairly balanced between genders, though one may have a slightly higher percentage of tenured professors.")
```
