```python
# ------------------------------
# Q1, Q2, and Q3: Probability & Hypothesis Testing
# ------------------------------

# Import necessary libraries
import pandas as pd
import numpy as np
from scipy import stats

# ------------------------------
# Load the dataset
# ------------------------------
ratings = pd.read_csv("student_rating_dataset_100.csv")
print("Dataset loaded successfully!\n")
print(ratings.head())
```

```python
# ------------------------------
# Q1: Probability of evaluation score > 4.5
# ------------------------------
p_greater_4_5 = (ratings['evaluation_score'] > 4.5).mean()
print(f"Q1. Probability of receiving evaluation score > 4.5 = {p_greater_4_5:.2f} ({p_greater_4_5*100:.1f}%)")
```

```python
# ------------------------------
# Q2: Probability of 3.5 < evaluation_score < 4.2
# ------------------------------
p_between_3_5_4_2 = ((ratings['evaluation_score'] > 3.5) & (ratings['evaluation_score'] < 4.2)).mean()
print(f"Q2. Probability of receiving evaluation score between 3.5 and 4.2 = {p_between_3_5_4_2:.2f} ({p_between_3_5_4_2*100:.1f}%)")
```

```python
# ------------------------------
# Q3: Two-tailed Z-test
# ------------------------------
# Given values
population_mean = 12        # μ (Historic mean of pro players)
population_std = 5.5        # σ (Standard deviation of pros)
sample_mean = 10.7          # x̄ (Regional players mean)
sample_size = 36            # n

# Step 1: Calculate Standard Error (SE)
SE = population_std / np.sqrt(sample_size)

# Step 2: Compute Z-score
z_score = (sample_mean - population_mean) / SE


# Step 3: Compute p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print("\nQ3. Two-tailed Z-test Results:")
print(f"Z-score = {z_score:.3f}")
print(f"P-value = {p_value:.4f}")

# Step 4: Decision at α = 0.05
alpha = 0.05
if p_value < alpha:
    print("❌ Reject Null Hypothesis (Means are significantly different).")
else:
    print("✅ Fail to Reject Null Hypothesis (No significant difference in means).")

# ------------------------------
# Hypothesis Summary
# ------------------------------
print("\nNull Hypothesis (H₀): Mean point of regional players = 12 (No difference).")
print("Alternative Hypothesis (H₁): Mean point of regional players ≠ 12 (There is a difference).")
```
